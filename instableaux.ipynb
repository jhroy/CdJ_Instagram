```python
import pandas as pan
import numpy as np
```

```python
mots = pan.read_csv("insta-motsseuls.csv", low_memory=False, names=["mot","interactions"])
```

```python
mots
```

```python
motstable = pan.pivot_table(mots,index=["mot"],values=["mot","interactions"],aggfunc=[len,np.sum])
```

```python
motstable
```

```python
motstable.sort_values(("len","interactions"), ascending=False).head(10000)
```

```python
motstable.sort_values(("len","interactions"), ascending=False).head(10000).to_csv("motsseuls-tableaucroise.csv")
```
